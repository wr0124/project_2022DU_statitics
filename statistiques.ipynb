### 4 Chi2 et mosaic plot

Pearson’s chi-squared test is a hypothesis test that is used to determine whether there is a significant association between two categorical variables in the data. The test involves two hypotheses (H0 & H1):

H0 : The two categorical variables have no relationship (independent)

H1 : There is a relationship (dependent) between two categorical variables

So as a null hypothesis, we keep the positive aspect of the test and in the alternate hypothesis, we keep the negative aspect. The positive aspect of chi-square is that there should not be any correlation because correlation can result in overfitting of the machine learning algorithm. The negative is that there is a correlation between the two categorical columns.

```python
mooc2.groupby(['Gender'])['HDI'].value_counts()
```

```python
table_chi2_temp={ 'HDI':['H','I','M'],
                 'Femme':[2546,233,147],
                 'Homme':[4716,883,432]
                }

table_chi2=pd.DataFrame(table_chi2_temp)
table_chi2['Row_total']=table_chi2[['Femme','Homme']].sum(axis=1)

print( table_chi2[['Femme','Homme','Row_total']].sum(axis=0) )

table_chi2.head()
```

```python
df2 = {'HDI': 'Col_total', 'Femme': 2926, 'Homme': 6031,'Row_total':8957}
table_chi2 = table_chi2.append(df2, ignore_index = True)
table_chi2.head()
```

```python
from scipy.stats import chi2_contingency
contingency_table = table_chi2[['Femme','Homme']]
contingency_table = np.array(contingency_table.drop([2]))
print(contingency_table)
```

```python
# hypothesis test of independence of the observed frequencies 
# in the contingency table
# Expected = row total x column total / all cell total
chi2,p_chi2,df,expected = chi2_contingency(contingency_table)
print(chi2,p_chi2,df,expected)
```

```python
print("p-value is   "+ str(p_chi2))
```

The probability that we would observe this or an even more imbalanced ratio by chance is about 5.755281578703447e-18 for the chi-square test. A commonly used significance level is 5% or 1%. if we adopt that, we can therefore conclude that our observed imbalance is statistically significant: significant effect of HDI on gender is present. The conlusion holds for the particular values in the table. 

```python
table1_chi2_temp={'Femme':[2546,233,147,2926], 
                  'Homme':[4716,883,432,6031],
                  'Row_total':[7262,1116,579,8957]
                 }
table1_chi2=pd.DataFrame(table1_chi2_temp)
table1_chi2=pd.DataFrame(table1_chi2_temp,index=['H','I','M','Col_total'])
table1_chi2.head()
```

```python
import numpy as np
expected=np.outer(table1_chi2['Row_total'][0:3],table1_chi2.loc['Col_total'][0:2])/8957
expected=pd.DataFrame(expected)
expected.columns=['Femme','Homme']
expected.index=['H','I','M']
expected
```

```python
observed=table1_chi2.iloc[0:3,0:2]
observed
```

```python
chi_squared_stat=(((observed-expected)**2)/expected).sum().sum()
print(chi_squared_stat)
```

```python
import scipy.stats as stats
crit=stats.chi2.ppf(q=0.95,df=2)
print('Critical value  '+str(crit))
```

```python
p_value=1-stats.chi2.cdf(x=chi_squared_stat,df=2)
print('p-value  '+str(p_value))
```
